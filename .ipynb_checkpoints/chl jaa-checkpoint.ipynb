```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Dell\Desktop\Engage\final.csv')
```

```python
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df['Number_of_Airbags']=df['Number_of_Airbags'].fillna(0)
```

```python
df.info()
```

```python
df['Basic_Warranty']= df['Basic_Warranty'].str.replace('2 Years / 100,000 Kms ', '2 years /100000 Kms ')
df['Basic_Warranty']= df['Basic_Warranty'].str.replace('8 yrs/1.6L kmBattery Warranty', '8 years /160000 kms ')

print(df['Basic_Warranty'].unique())
```

```python
df['Extended_Warranty']=df['Extended_Warranty'].replace('3 Years/Unlimited km Warranty*', '3 Years Unlimited Kilometres')
```

```python
print(df['Extended_Warranty'].unique())
```

```python
pd.set_option('display.max_columns', None)

df1=pd.get_dummies(df,columns=df.iloc[:,[0,1,7,9,10,22,25,26,30,31,34,35,38,39,40,41,42,43,44]].columns.to_list())

#used filtration
```

```python
df1.head()
```

```python
df1.drop(columns= ['Variant','Power','Torque','Type','Odometer','Speedometer','Tachometer','Audiosystem','Aux-in_Compatibility','12v_Power_Outlet','Boot-lid_Opener','Fuel-lid_Opener', 'Fuel_Gauge','Airbags','Make_Aston Martin','Model_3-Series','Emission_Norm_BS 6', 'Fuel_Type_CNG','Body_Type_Convertible','Start_/_Stop_Button_Not known','Basic_Warranty_1 year without mileage limit','Bluetooth_Not known','Child_Safety_Locks_Not known','Extended_Warranty_1 Years Unlimited Kilometres','Handbrake_Automatic','Hill_Assist_Not known', 'Parking_Assistance_Front & rear sensors with 360 degree view','Infotainment_Screen_Not known','Navigation_System_Not known', 'Third_Row_AC_Vents_Not Applicable', 'Second_Row_AC_Vents_Not known','Engine_Type_1.2 L Revotron (BS VI)','Rain_Sensing_Wipers_Not known'],inplace=True)                  
```

```python
df1.drop(columns=['Ground_Clearance'],inplace=True)
df1.head()
```

```python
X = df1.drop(columns=['Ex-Showroom_Price'])
y= np.log(df1['Ex-Showroom_Price'])
#filteration+1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15, random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
```

```python
X_train.isnull().sum()
X_train.drop(columns=['Boot_Space'],inplace=True)
X_train.isnull().sum()
```

```python
X_test.drop(columns=['Boot_Space'],inplace=True)
```

```python
lr= LinearRegression()

lr.fit(X_train, y_train)

y_pred= lr.predict(X_test) 
print('R2 score', r2_score(y_test, y_pred)) 
print('MAE', mean_absolute_error(y_test, y_pred))
```

```python
# we will we using ridge regression
ridge = Ridge(alpha=10)
ridge.fit(X_train,y_train)

y_pred = ridge.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
lasso = Lasso(alpha=0.001)
lasso.fit(X_train,y_train)

y_pred = lasso.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(X_train,y_train)

y_pred = knr.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
decisiontree = DecisionTreeRegressor(max_depth=8)
decisiontree.fit(X_train,y_train)

y_pred = decisiontree.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
svr = SVR(kernel='rbf',C=10000,epsilon=0.1)
svr.fit(X_train,y_train)

y_pred = svr.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
randomforest = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

randomforest.fit(X_train,y_train)

y_pred = randomforest.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
extratree = ExtraTreesRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

extratree.fit(X_train,y_train)

y_pred = extratree.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
adaboost = AdaBoostRegressor(n_estimators=15,learning_rate=1.0)
adaboost.fit(X_train,y_train)

y_pred = adaboost.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
gradientboost = GradientBoostingRegressor(n_estimators=500)

gradientboost.fit(X_train,y_train)

y_pred = gradientboost.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
